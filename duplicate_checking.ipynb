# cancer disease duplicate data remove and merging two cancer dataset

```python
df1 = pd.read_csv('pubmed_ALL_cancer_environment_20251214_010208.csv', header=0, index_col=0)
```

```python
df2 = pd.read_csv('cancer.csv', header=0, index_col=0)
```

```python
df = pd.concat([df1, df2], ignore_index=True)
```

```python
df.head()
```

```python
df.shape
```

```python
def clean_text(text):
    if pd.isna(text):
        return ""
    return (
        str(text)
        .lower()
        .strip()
    )

df['title_clean'] = df['Title'].apply(clean_text)
df['abstract_clean'] = df['Abstract'].apply(clean_text)
```

```python
duplicates = df[
    df.duplicated(subset=['title_clean'], keep=False) |
    df.duplicated(subset=['abstract_clean'], keep=False)
]

print("Total duplicates found:", duplicates.shape[0])
```

```python
duplicates = df[
    df.duplicated(subset=['title_clean'], keep=False) |
    df.duplicated(subset=['abstract_clean'], keep=False)
]

print("Total duplicates found:", duplicates.shape[0])
```

```python
df_deduplicated = df.loc[
    ~(
        df.duplicated(subset=['title_clean'], keep='first') |
        df.duplicated(subset=['abstract_clean'], keep='first')
    )
]
```

```python
print("Before:", df.shape)
print("After:", df_deduplicated.shape)
```

# *LUNG* disease duplicate data remove and merging two cancer dataset

```python
df_lung = pd.read_csv("lung.csv", header=0, index_col=0)
```

```python
df2_lung = pd.read_csv("pubmed_ALL_lung_environment_20251214_011723.csv", header=0, index_col=0)
```

```python
df_lung = pd.concat([df_lung, df2_lung], ignore_index=True)
```

```python
df_lung.head()
```

```python
df_lung.shape
```

```python

def clean_text(text):
    if pd.isna(text):
        return ""
    return str(text).lower().strip()

df_lung['title_clean'] = df_lung['Title'].apply(clean_text)
df_lung['abstract_clean'] = df_lung['Abstract'].apply(clean_text)

duplicates = df_lung[
   df_lung.duplicated(subset=['title_clean'], keep=False) |
    df_lung.duplicated(subset=['abstract_clean'], keep=False)
]

print("Total duplicate rows found:", duplicates.shape[0])

df_lung_deduplicated = df_lung.loc[
    ~(
        df_lung.duplicated(subset=['title_clean'], keep='first') |
        df_lung.duplicated(subset=['abstract_clean'], keep='first')
    )
]

print("Rows before deduplication:", df_lung.shape[0])
print("Rows after deduplication:", df_lung_deduplicated.shape[0])

df_lung_deduplicated = df_lung_deduplicated.drop(
    columns=['title_clean', 'abstract_clean']
)



df_deduplicated.head()
```
